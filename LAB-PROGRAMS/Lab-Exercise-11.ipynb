```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('mxmh_survey_results.csv')
df.head(10)
```

```python
df.corr()
```

```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr)
plt.show()
```

```python
x = df["Fav genre"][0:20]
y = df["Hours per day"][0:20]
plt.scatter(x,y)
plt.show()
```

```python
x = df["Fav genre"][0:5]
y = df["Hours per day"][0:5]
plt.xlabel('Fav genre')
plt.ylabel('Hours per day')
plt.title('Music Streaming')
plt.bar(x, y)
plt.show()
```

```python
plot = df["Hours per day"].value_counts().plot(kind="bar")
plt.xlabel('Fav genre')
plt.ylabel('Primary streaming service')
plt.title('Music Streaming Service')
plt.show()
```

```python
plot = df["Hours per day"].value_counts().plot(kind="line")
plt.ylabel("")
plt.title('Music Streaming Service')
plt.show()
```
